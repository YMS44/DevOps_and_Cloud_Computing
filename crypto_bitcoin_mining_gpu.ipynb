```python
from cryptography.hazmat.primitives import hashes
import time
```

```python
nonce = 0
minzeroes = 2
```

```python
start_time = time.time()
```

```python
while True:
  digest  = hashes.Hash(hashes.SHA256())
  digest.update(b"Alice->Bob, 5 BTC  Carol->Dave, 2 BTC ...")
  s = bytearray(str(nonce),encoding='ISO-8859-1')
  digest.update(s)
  b = digest.finalize()
  h = b.hex()
  print("Iteration: %3d" %nonce , "Hash: %s" %h)
  nonce += 1
  if str(h)[:minzeroes]=='0'*minzeroes:
    print(f"Hash starting with at least {minzeroes} zeros found!!!")
    break
```

```python
end_time = time.time()
elapsed_time = end_time - start_time
print("Elapsed time:", elapsed_time, "seconds")
```
